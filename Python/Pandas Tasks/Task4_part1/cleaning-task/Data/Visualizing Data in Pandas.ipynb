# Pandas Visualization

![Types of data](Types-of-Data-Infographic.png)

- **Categorical Data** → Bar Chart, Pie Chart  
- **Numerical Data** → Histogram, Box Plot  
- **Time Series Data** → Line Chart  
- **Relational Data** → Scatter Plot, Heatmap  

1. **Nominal Data** (e.g., Gender, Hair Color) → **Bar Chart / Pie Chart**
2. **Ordinal Data** (e.g., Letter Grades, Economic Status) → **Bar Chart / Ordered Column Chart**
3. **Discrete Data** (e.g., Number of Students) → **Bar Chart**
4. **Continuous Data** (e.g., Height, Speed) → **Histogram / Line Chart / Scatter Plot**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('classic')
```

```python
df.plot()
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores', subplots=True)
```

```python
df.plot(kind = 'bar', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores', figsize = (10,5))
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'Yellow')
```

```python
df.plot.hist(bins = 10)
```

```python
df2 = pd.DataFrame(np.random.randint(1, 7, 6000), columns=['one'])
df2['two'] = df2['one'] + np.random.randint(1, 7, 6000)
ax = df2.plot.hist(bins=12, alpha=0.5)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y='Flavor Rating',figsize=(10,10))
```
