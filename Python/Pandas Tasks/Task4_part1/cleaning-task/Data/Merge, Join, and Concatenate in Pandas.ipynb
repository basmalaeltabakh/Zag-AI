# Merge, Join, and Concatenate 

![image.png](join.png)

```python
import pandas as pd
```

```python
df1 = pd.read_csv("LOTR.csv")
df1
```

```python
df2 = pd.read_csv("LOTR 2.csv")
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID'])  # x for df1 and y for df2
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2)
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left',rsuffix = '_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right', how = 'outer')
df4
```

```python
pd.concat([df1,df2]) # half of data in df1 and the other half of data in df2
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)  # df1 is diff from df2 but still df2 is part of df1
```

```python
df1.append(df2)
```
