# Loading Dataset

```python
df = sns.load_dataset('tips')
df.head()
```

# Distribution Plots 

# Correlation Heatmap

```python
numeric_cols = ['total_bill','tip','size']
df[numeric_cols].corr()
```

```python
sns.heatmap(df[numeric_cols].corr())
```

```python
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="YlGnBu")
```

# Joint Plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.jointplot(x="tip", y="total_bill", data=df, kind="kde", fill=True, cmap="YlGnBu")
```

# Pair Plot 

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

# Dist Plot

```python
sns.distplot(df['tip'])
```

```python
sns.histplot(df['tip'], kde=True, bins=15)
```

# Categorical Plots

# Count Plot

```python
sns.countplot(x='sex', data=df, palette='viridis')
```

```python
sns.countplot(x='smoker',data=df, palette='viridis')
```

```python
sns.countplot(x='day',data=df, palette='viridis')
```

### Horizontal Count Plot

```python
sns.countplot(y='sex',data=df, palette='viridis')
```

# Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=df, palette='viridis')
```

```python
sns.barplot(x='smoker',y='total_bill',data=df, palette='viridis')
```

# Box Plot

```python
sns.boxplot(x='smoker',y='total_bill',data=df, palette='viridis')
```

```python
sns.boxplot(x='sex',y='total_bill',data=df, palette='viridis')
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v', palette='viridis')
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=df, palette='viridis')
```

# Violin Plot

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```

# Scatter Plot

```python
sns.scatterplot(x="tip", y="total_bill", data=df, hue="day", size="size", palette="YlGnBu")
```

```python
sns.stripplot(x="day", y="tip", data=df, hue="sex", dodge=True, palette="YlGnBu")
```

```python
sns.barplot(x="day", y="tip", data=df, hue="sex", palette='viridis')
```
