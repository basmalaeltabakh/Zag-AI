### **1.Explore Data**

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#To take a look at the first five rows as default
df = pd.read_csv("crime_data.csv")
df.head(5)
```

```python
#To take a look at the last five rows as default
df.tail(5)
```

```python
#print the summary of the DataFrame
df.info()
```

```python
#Print the total number of rows and columns in the data set
print(f"Total Rows: {df.shape[0]}, Total Columns: {df.shape[1]}")
```

```python
# Shows basic statistics for each numerical column in the DataFrame
df.describe()
```

```python
# Check for nan values if found then drop them
# removing Nan Value or missing values
df.isnull().sum()
df.dropna(inplace=True)
```

```python
# Same for duplicates check for them first then drop if found any
df.duplicated().sum() 
df.drop_duplicates(inplace=True)  
print(f"Duplicate Rows: {df.duplicated().sum()}") 
```

```python
#Count unique values
df.nunique()
```

### **2.Find the 5 states with the lowest and highest `crime rates`**

```python
# find the 5 states with the lowest crime rates
lowest_crime_states = df.nsmallest(5, "Crime Rate")[["State", "Crime Rate"]]
lowest_crime_states
```

```python
# Create a bar plot to visualize the 5 states with the lowest crime rate
plt.figure(figsize=(10, 5))
sns.barplot(data=lowest_crime_states, x="State", y="Crime Rate", palette="Blues")

# Set title and labels for the plot
plt.title("Top 5 States with Lowest Crime Rate")
plt.xlabel("State")
plt.ylabel("Crime Rate")

# Display the plot
plt.show()

# Print your conclusions
print("Conclusion: The states with the lowest crime rates have better security measures or lower population densities, affecting the crime rate positively.")
```

```python
# get the 5 states with the highest crime rates.
highest_crime_states = df.nlargest(5, 'Crime Rate')[['State', 'Crime Rate']]
highest_crime_states 
```

```python
# Create a bar plot to visualize the 5 states with the Highest crime rate
plt.figure(figsize=(10, 5))
sns.barplot(x='Crime Rate', y='State', data=highest_crime_states, palette="Reds_r")

# Set title and labels for the plot

plt.title("5 States with Highest Crime Rates")
plt.xlabel("Crime Rate")
plt.ylabel("State")

# Display the plot

plt.show()
# Print your conclusions
print("Conclusion: The states with the highest crime rates may have larger populations or higher crime-prone areas, leading to increased crime rates.")
```

### **3.Find the sum of `Rape Cases` for each year**

```python
# Group the dataset by 'Year' and calculate the sum of 'Rape Cases' for each year
rape_cases_per_year = df.groupby('Year')['Rape Cases'].sum()

# Plot the line chart showing the trend of total rape cases per year
plt.figure(figsize=(8,5))
sns.lineplot(x=rape_cases_per_year.index, y=rape_cases_per_year.values, marker="o")

# Title and labels for the plot
plt.title("Total Rape Cases per Year")
plt.xlabel("Year")
plt.ylabel("Total Rape Cases")
plt.grid(True)

# Show the plot
plt.show()

# Print your conclusions from this plot
print("Conclusions: The trend in rape cases over the years helps identify whether crime is increasing or decreasing and highlights necessary preventive measures.")
```

### **4.Find the correlation between `total crimes` and `population`**

```python
# Create a scatter plot with Population on the x-axis and Total Crimes on the y-axis
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Population"], y=df["Total Crimes"])

# Set title and labels for the plot
plt.title("Population vs. Total Crimes")
plt.xlabel("Population")
plt.ylabel("Total Crimes")

# Show the plot
plt.show()

# Print your conclusions from this plot
print("Conclusions: A positive correlation might indicate that more populated areas experience higher crime rates.")
```

### **5.Find out how the average of `crime clearance` rate vary across different `states`**

```python
# Group the data by 'State' and calculate the mean 'Crime Clearance Rate' for each state, then sort them in descending order
avg_clearance_rate = df.groupby('State')['Crime Clearance Rate'].mean().sort_values(ascending=False)

# Create a bar plot using the calculated average crime clearance rates
plt.figure(figsize=(12, 6))  
sns.barplot(x=avg_clearance_rate.index, y=avg_clearance_rate.values, palette="coolwarm")

# Set the title and labels for the plot
plt.title("Average Crime Clearance Rate by State")
plt.xlabel("State")
plt.ylabel("Crime Clearance Rate (%)")

# Rotate x-axis labels (state names) to avoid overlap and ensure readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
# Print your conclusions from this plot


print("Conclusions:")
print("- Some states have significantly higher crime clearance rates, indicating better law enforcement efficiency.")
print("- States with lower clearance rates may need policy improvements or additional resources to solve crimes effectively.")
```

### **6.Find the Outliers of `abduction cases` in `western states`**

Note: wester states are 'Goa', 'Gujarat', 'Maharashtra', 'Rajasthan'

```python
# Defining the western states
western_states = ['Goa', 'Gujarat', 'Maharashtra', 'Rajasthan']


# Filtering the dataset for western states
df_western = df[df['State'].isin(western_states)]

# Plot the boxplot for Abduction Cases in Western States
plt.figure(figsize=(8,5))
sns.boxplot(x="State", y="Abduction Cases", data=df_western)


# Set the title and labels for the plot
plt.title("Outliers in Abduction Cases in Western States")
plt.xlabel("State")
plt.ylabel("Abduction Cases")

# Show the plot
plt.show()
# Print your conclusions from this plot

print("Conclusions: Outliers indicate states with extreme cases of abduction, requiring further investigation.")
```

### **7.Find the trend in the number of `murder cases` over the `years`**

```python
# get the sum of the murder cases over the years
murder_cases_per_year = df.groupby("Year")["Murder Cases"].sum()

# Create a line plot with murder cases trend
plt.figure(figsize=(8,5))
sns.lineplot(x=murder_cases_per_year.index, y=murder_cases_per_year.values, marker="o")

# Set the title and labels for the plot
plt.title("Trend of Murder Cases Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Murder Cases")
plt.grid(True)

# Show the plot
plt.show()
# Print your conclusions from this plot
print("Conclusions: The murder trend over the years shows whether homicide cases are increasing or decreasing.")
```

### **8.Find out which `state` has highest number of `arrest`**

```python
#pick the top 10 states
top_arrest_states = df.groupby("State")["Arrests"].sum().nlargest(10)

#barplot for the top 10 states
plt.figure(figsize=(10,5))
sns.barplot(x=top_arrest_states.values, y=top_arrest_states.index, palette="coolwarm")

# Set the title and labels
plt.title("Top 10 States with Highest Number of Arrests")
plt.xlabel("Number of Arrests")
plt.ylabel("State")

# Show the plot
plt.show()

# Print your conclusions from this plot

print("Conclusions: Higher arrests indicate strong law enforcement or higher crime rates.")
```

### **9.Find the average `crime rate` across southern states over the `years`**

Note: southern states are 'Andhra Pradesh', 'Telangana', 'Tamil Nadu', 'Karnataka', 'Kerala', 'Puducherry'

```python
#Southern States
southern_states = ['Andhra Pradesh', 'Telangana', 'Tamil Nadu', 'Karnataka', 'Kerala', 'Puducherry']
# Filter the dataset for Southern States
df_southern = df[df['State'].isin(southern_states)]


# Calculate the average crime rate by state and year
avg_crime_rate_south = df_southern.groupby("Year")["Crime Rate"].mean()

# Plot the line plot for Southern States
plt.figure(figsize=(8,5))
sns.lineplot(x=avg_crime_rate_south.index, y=avg_crime_rate_south.values, marker="o")

# Set title and labels
plt.title("Average Crime Rate in Southern States Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Crime Rate")
plt.grid(True)

# Show the plot
plt.show()

# Print your conclusions from this plot
print("Conclusions: Southern states may have different law enforcement effectiveness and crime trends.")

```

### **10.Find the Total number of `cases` for each crime type**

Note: cases are Rape cases, Abduction cases, Murder cases

```python
# Sum the cases for each crime type
crime_type_counts = df[['Rape Cases', 'Abduction Cases', 'Murder Cases']].sum()

# Plot the donut chart

plt.figure(figsize=(7, 7))
colors = ['#ff9999', '#66b3ff', '#99ff99'] 
plt.pie(crime_type_counts, labels=crime_type_counts.index, autopct='%1.1f%%', colors=colors, wedgeprops={'edgecolor': 'black'})
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)
# Set the title
plt.title("Total Number of Cases for Each Crime Type")

# Show the plot
plt.show()

# Print your conclusions from this plot
print("conclusions: ")
print("- The pie chart shows the proportion of different crime types in the dataset.")
print("- The crime type with the highest number of cases is the most prevalent in the dataset.")
print("- This visualization helps in understanding which type of crime is more common.")
```

### **12.Find the Correlation Matrix of Crime Data**

```python
# Filter data to ensure only numeric data is left
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Set up the visual style of the plot
plt.figure(figsize=(10, 6))
sns.set(style="white")

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Set the title
plt.title("Correlation Matrix of Crime Data")

# Display the plot
plt.show()
# Print your conclusions from this plot
print("conclusions: ")
print("- Darker red/blue colors indicate stronger positive/negative correlations.")
print("- High correlation between some crime types suggests they may be related.")
print("- This analysis helps understand which crimes often occur together.")
```
